### Manual Input

```python
import pandas as pd
import numpy as np
from keras.models import load_model
from sklearn.preprocessing import StandardScaler

# Load the saved model
model = load_model('LTSM_model.h5')

# Load the new data as a Pandas DataFrame with column names
new_data = pd.DataFrame({'Time_Sec':[1440],
              'diff_PT':[-10],
              'diff_AST':[-5],
              'diff_RB':[1],
              'diff_ORB':[4],
              'diff_DRB':[-3],
              'diff_ST':[-2],
              'diff_BLK':[-2],
              'diff_TO':[1],
              'diff_PF':[-2],
              'diff_Tech':[0],
              'diff_PTA':[4],
              'Home_Leading':[0]})

# Scale and normalize the new data using the same StandardScaler instance used to train the model
scaler = StandardScaler()
X_new = scaler.fit_transform(new_data)

# Reshape the input data to fit the expected shape for an LSTM model
X_new = np.reshape(X_new, (X_new.shape[0], 1, X_new.shape[1]))

# Make predictions on the new data using the loaded model
y_pred = model.predict(X_new)

# Print the predictions
print(y_pred)
```

### Using Application Data (Game is already finished)

```python
import pandas as pd
import numpy as np
from keras.models import load_model
from sklearn.preprocessing import StandardScaler

# Load the saved model
model = load_model('LTSM_model.h5')

# Load the new data as a Pandas DataFrame with column names
new_data = pd.read_csv('Application_Data.csv')

# Scale and normalize the new data using the same StandardScaler instance used to train the model
scaler = StandardScaler()
X_new = scaler.fit_transform(new_data)

# Reshape the input data to fit the expected shape for an LSTM model
X_new = np.reshape(X_new, (X_new.shape[0], 1, X_new.shape[1]))

# Make predictions on the new data using the loaded model
y_pred = model.predict(X_new)

# Print the predictions
print(y_pred)
```

```python
import matplotlib.pyplot as plt

# assuming y_pred is a 1D array of predicted probabilities for each time step
# and X is a 2D array of input data with a column named Time_sec for the time points
plt.plot(new_data['Time_Sec'],y_pred)  # assuming Time_sec is the first column of X
plt.xlabel('Time (seconds)')
plt.ylabel('Predicted Probability of Home Team Winning')
plt.show()
```
