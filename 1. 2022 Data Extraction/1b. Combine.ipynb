```python
import pandas as pd
import numpy as np
```

```python
ATL=pd.read_csv('ATL-2022.csv')
BOS=pd.read_csv('BOS-2022.csv')
BRK=pd.read_csv('BRK-2022.csv')
CHI=pd.read_csv('CHI-2022.csv')
CHO=pd.read_csv('CHO-2022.csv')
CLE=pd.read_csv('CLE-2022.csv')
DAL=pd.read_csv('DAL-2022.csv')
DEN=pd.read_csv('DEN-2022.csv')
DET=pd.read_csv('DET-2022.csv')
GSW=pd.read_csv('GSW-2022.csv')
HOU=pd.read_csv('HOU-2022.csv')
IND=pd.read_csv('IND-2022.csv')
LAC=pd.read_csv('LAC-2022.csv')
LAL=pd.read_csv('LAL-2022.csv')
MEM=pd.read_csv('MEM-2022.csv')
MIA=pd.read_csv('MIA-2022.csv')
MIL=pd.read_csv('MIL-2022.csv')
MIN=pd.read_csv('MIN-2022.csv')
NOP=pd.read_csv('NOP-2022.csv')
NYK=pd.read_csv('NYK-2022.csv')
OKC=pd.read_csv('OKC-2022.csv')
ORL=pd.read_csv('ORL-2022.csv')
PHI=pd.read_csv('PHI-2022.csv')
PHO=pd.read_csv('PHO-2022.csv')
POR=pd.read_csv('POR-2022.csv')
SAC=pd.read_csv('SAC-2022.csv')
SAS=pd.read_csv('SAS-2022.csv')
TOR=pd.read_csv('TOR-2022.csv')
UTA=pd.read_csv('UTA-2022.csv')
WAS=pd.read_csv('WAS-2022.csv')
```

```python
merged_df = pd.concat([ATL,BOS,BRK,CHI,CHO,CLE,DAL,DEN,DET,GSW,HOU,IND,LAC,LAL,MEM,MIA,MIL,MIN,NOP,NYK,OKC,ORL,PHI,PHO,POR,SAC,SAS,TOR,UTA,WAS], axis=0, ignore_index=True)
```

```python
merged_df=merged_df.drop_duplicates()
```

```python
merged_df['Result']=""
```

```python
def Result(merged_df):
    if merged_df['Time_48']=='0 days 00:48:00.000000000':
        if merged_df['Team_A_Score']>merged_df['Team_B_Score']:
            return 'Team A'
        elif merged_df['Team_A_Score']==merged_df['Team_B_Score']:
            pass
        else:
            return 'Team B'
    elif merged_df['Time_48']=='0 days 00:53:00.000000000':
        if merged_df['Team_A_Score']>merged_df['Team_B_Score']:
            return 'Team A'
        elif merged_df['Team_A_Score']==merged_df['Team_B_Score']:
            pass
        else:
            return 'Team B'
    elif merged_df['Time_48']=='0 days 00:58:00.000000000':
        if merged_df['Team_A_Score']>merged_df['Team_B_Score']:
            return 'Team A'
        elif merged_df['Team_A_Score']==merged_df['Team_B_Score']:
            pass
        else:
            return 'Team B'
    elif merged_df['Time_48']=='0 days 01:03:00.000000000':
        if merged_df['Team_A_Score']>merged_df['Team_B_Score']:
            return 'Team A'
        elif merged_df['Team_A_Score']==merged_df['Team_B_Score']:
            pass
        else:
            return 'Team B'    
    else:
        pass
```

```python
merged_df['Result'] = merged_df.apply(Result, axis=1)
```

```python
merged_df['Result']=merged_df['Result'].bfill()
```
